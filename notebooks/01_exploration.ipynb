```python
# Import moduls
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import random
import re 
from types import NoneType
from xmlrpc.client import Boolean
import cv2
import torchvision.transforms as transforms

import os

pd.set_option('display.max_columns', None)

from core.data import load_from_kaggle
```

Load data and save them in the data/raw folder. 

```python
dataset_link = "yusufmurtaza01/chest-xray-pneumonia-balanced-dataset"
destination = "../data/raw"
dataset_name = dataset_link.split("/")[-1]

files = load_from_kaggle(
    dataset_link=dataset_link, 
    destination=destination,
    )
```

```python
DATASET_NAME = dataset_name
# das speichern
```

```python
def is_augmented(name: str) -> Boolean:
    REGULAR_EXPRESSION = 'aug'
    result = re.search(REGULAR_EXPRESSION, name)
    return not isinstance(result, NoneType)

def get_random_item(path_folder: str) -> str:
    files = os.listdir(path_folder)
    random_number = random.randint(0, len(files)-1)
    return files[random_number]

def get_tensor_image(set_name: str, label: str, allowed_augmented: Boolean) -> tensor:
    path_folder = "/".join(["../data/raw/", DATASET_NAME, set_name, label])
    file_name = get_random_item(path_folder)
    if not allowed_augmented: 
        while is_augmented(file_name):
            file_name = get_random_item(path_folder)
    path_file = "/".join(["../data/raw/", DATASET_NAME, set_name, label, file_name])
    image = cv2.imread(path_file)
    transform = transforms.Compose([transforms.ToTensor()])
    tensor = transform(image)
    return tensor[0,:,:]
```

Plot some of the non-augmented images.

```python
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(10,10))
cols, rows = 3, 3
set_name = 'train'
for i in range(3):
    if i == 0:
        label = 'NORMAL'
    else: 
        label = 'PNEUMONIA'
    for j in range(3):
        tensor = get_tensor_image(set_name, label, False)
        ax[i,j].set_title(label)
        ax[i,j].imshow(tensor.squeeze(), cmap="gray")
        ax[i,j].spines['bottom'].set_visible(False)
        ax[i,j].spines['top'].set_visible(False)
        ax[i,j].spines['right'].set_visible(False)
        ax[i,j].spines['left'].set_visible(False)
        ax[i,j].tick_params(axis='both', which='both', bottom=False, top=False, 
                            labelbottom=False, left=False, right=False, labelleft=False)
       
```
