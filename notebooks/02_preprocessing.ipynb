```python
dict_label_decoding = {0: 'NORMAL', 1: 'PNEUMONIA'}

def plot_first_train_images(resolution: int) -> None:
    training_data = CustomImageDataset('train', transform=True, target_transform=Lambda(lambda y: torch.zeros(2, dtype=torch.float).scatter_(0, torch.tensor(y), value=1)), resolution = resolution)
    fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(10,10))
    cols, rows = 3, 3
    set_name = 'train'
    for i in range(3):
        for j in range(3):
            tensor =  training_data[j + i*3][0]
            label_int = training_data[j + i*3][1].argmax(0).item()
            label = dict_label_decoding[label_int]
            ax[i,j].set_title(label + ' \n with resolution ' + str(resolution))
            ax[i,j].imshow(tensor.squeeze(), cmap="gray")
            ax[i,j].spines['bottom'].set_visible(False)
            ax[i,j].spines['top'].set_visible(False)
            ax[i,j].spines['right'].set_visible(False)
            ax[i,j].spines['left'].set_visible(False)
            ax[i,j].tick_params(axis='both', which='both', bottom=False, top=False, 
                            labelbottom=False, left=False, right=False, labelleft=False)
```

```python
for resolution in [10,50,100,500,1000]:
    plot_first_train_images(resolution)
```
